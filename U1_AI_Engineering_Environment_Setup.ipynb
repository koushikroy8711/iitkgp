```python
# 1. Create a tensor representing the number 2.0.
# Setting 'requires_grad=True' is the key here—it tells PyTorch's
# engine to track every mathematical operation done to 'x' so it
# can calculate the gradient (derivative) later.
x = torch.tensor(2.0, requires_grad=True)

# 2. Define the equation.
# Here, y is equal to x squared.
y = x**2

# 3. Trigger the automatic differentiation.
# Calling backward() computes the derivative of 'y' with respect to 'x'.
# Mathematically, the derivative of x^2 is 2x.
y.backward()

# 4. Print the result.
# PyTorch stores the computed derivative inside the '.grad' attribute of x.
# Since x = 2.0, the calculation is 2 * (2.0) = 4.0.
print(x.grad)
```
